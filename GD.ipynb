# 1. Define Cost Function
# 2. Compute Gradient
# 3. Initialize Parameter
# 4. Update Rule
# 5. Track Progress
# 6. Visualize

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(w):
    return (w - 3) ** 2

# Define the gradient (derivative of cost function)
def gradient(w):
    return 2 * (w - 3)

# Gradient Descent Algorithm
def gradient_descent(initial_w, learning_rate, iterations):
    w = initial_w  # Initialize weight
    cost_values = []  # Store cost function values
    w_values = []  # Store weight values
    
    for i in range(iterations):
        grad = gradient(w)  # Compute gradient
        w = w - learning_rate * grad  # Update weight
        
        cost_values.append(cost_function(w))  # Store cost
        w_values.append(w)  # Store weight

        print(f"Iteration {i+1}: w = {w:.4f}, Cost = {cost_function(w):.4f}")

    return w_values, cost_values

# Parameters
initial_w = 10  # Starting point
learning_rate = 0.1
iterations = 20

# Run Gradient Descent
w_values, cost_values = gradient_descent(initial_w, learning_rate, iterations)

# Plot cost function vs iterations
plt.plot(range(1, iterations + 1), cost_values, marker='o')
plt.xlabel("Iterations")
plt.ylabel("Cost Function Value")
plt.title("Gradient Descent Optimization")
plt.show()
```
